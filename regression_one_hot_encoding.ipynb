```python
import pandas as pd
```

```python
dataset=pd.read_csv("/content/MBA.csv")
dataset
```

```python
dataset.isnull()
```

```python
dataset.isnull().sum()
```

```python
dataset_1=dataset.drop(columns=["race"])

dataset_1
```

```python
dataset_1.isnull().sum()
```

```python
dataset_2= dataset_1.drop(columns=['application_id', 'gpa','work_industry'])
dataset_2
```

```python
dataset_2["major"].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(handle_unknown='ignore')
dataset_3=pd.get_dummies(dataset_2,columns=["gender", "international", "major", "admission"],drop_first=True,dtype=int)
```

```python
dataset_2["international"].value_counts()
```

```python
dataset_3
```

```python
X=dataset_3[["work_exp","gender_Male","international_True",	"major_Humanities","major_STEM","admission_Rejected","admission_Waitlist"]]
y=dataset_3[["gmat"]]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train= scaler.fit_transform(X_train)
X_test= scaler.fit_transform(X_test)
```

```python
X_train
```

```python
from sklearn.svm import SVR

model = SVR()  # Create model object
model.fit(X_train, y_train)  # Fit on training data
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
```

```python
r2
```

```python
import pickle
filename="SupportVectorRegression.sav"
pickle.dump(model,open(filename,"wb"))
```
