```python
#import pandas
import pandas as pd
```

```python
#load the dataset
dataset=pd.read_csv("/content/MBA.csv")
dataset
```

```python
#drop the unwanted columns and load after the drop method is done
dataset_encoded=dataset.drop(columns=["race","gender","international"])
dataset_encoded
```

```python
#count the values in independent variable major
dataset_encoded["major"].value_counts()
```

```python
#count the values of work industry independent variable
dataset_encoded["work_industry"].value_counts()
```

```python
#import label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset_encoded['major_encoded'] = le.fit_transform(dataset['major'])
dataset_encoded['work_industry_encoded'] = le.fit_transform(dataset['work_industry'])
```

```python
# view the column after encoded
dataset_encoded["major_encoded"]
```

```python
#view the column after encoded
dataset_encoded["work_industry_encoded"]
```

```python
dataset_encoded
```

```python
dataset_encoded=dataset_encoded.drop(columns=["major","work_industry"])
```

```python
dataset_encoded
```

```python
#assign the dependent and independent variable (x,y)
X=dataset_encoded[["application_id","gpa","gmat","work_exp","major_encoded","work_industry_encoded"]]
y=dataset_encoded[["admission"]]
```

```python
# do  train ,test and split method
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
# import decision tree classifier to train the model
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
#import accuracy score to find a accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
#import pickle to save a file
import pickle
filename="DecisionTreeClassification.sav"
pickle.dump(model,open(filename,"wb"))
```

```python
# load the saved file
load_model=pickle.load(open("/content/DecisionTreeClassification.sav","rb"))
```
