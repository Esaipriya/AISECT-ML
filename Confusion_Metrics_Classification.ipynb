```python
#import pandas library
import pandas as pd
```

```python
#read the dataset using pandas
dataset=pd.read_csv("/content/personality_datasert.csv")
dataset
```

```python
#drop the unwanted columns
dataset=dataset.drop(columns=["Post_frequency"])
dataset
```

```python
#take the value counts for Stage_fear column
dataset["Stage_fear"].value_counts()
```

```python
#take the value counts for Drained_after_socializing column
dataset["Drained_after_socializing"].value_counts()
```

```python
#using info() method to know the informations like column names,dtype, non null count about the dataset
dataset.info()
```

```python
#import onehot encoding to encode the columns
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(handle_unknown='ignore')
dataset = pd.get_dummies(dataset, columns=['Stage_fear','Drained_after_socializing'],drop_first=True, dtype=int)
```

```python
dataset
```

```python
# change the float value into round or integer
dataset["Time_spent_Alone"] = dataset["Time_spent_Alone"].round().astype(int)
dataset["Social_event_attendance"] = dataset["Social_event_attendance"].round().astype(int)
dataset["Going_outside"] = dataset["Going_outside"].round().astype(int)
dataset["Friends_circle_size"] = dataset["Friends_circle_size"].round().astype(int)
dataset
```

```python
# assign the dependent and independent variable
X=dataset[["Time_spent_Alone","Social_event_attendance","Going_outside","Friends_circle_size","Stage_fear_Yes","Drained_after_socializing_Yes"]]
y=dataset[["Personality"]]
```

```python
# do train test split method
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
# import decision tree classifier to train the model
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
#calculate the confusion metrix values
#precision score
from sklearn.metrics import precision_score
precision_score(y_test, y_pred, average=None)
```

```python
#recall score
from sklearn.metrics import recall_score
recall_score(y_test, y_pred, average=None)
```

```python
#f1 score
from sklearn.metrics import f1_score
f1_score(y_test, y_pred, average=None)
```

```python
#import pickle to save a file
import pickle
filename="Confusion_metrics_Classification.sav"
pickle.dump(model,open(filename,"wb"))
```

```python
# load the saved file
load_model=pickle.load(open("/content/Confusion_metrics_Classification.sav","rb"))
```
