```python
import pandas as pd
```

```python
dataset=pd.read_csv("/content/house_price_regression_dataset.csv")
dataset
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset=dataset.drop(columns=["Neighborhood_Quality"])
```

```python
dataset["Num_Bathrooms"].value_counts()
```

```python
dataset["Num_Bedrooms"].value_counts()
```

```python
dataset["Year_Built"].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(handle_unknown='ignore')
dataset=pd.get_dummies(dataset,columns=["Num_Bedrooms","Num_Bathrooms"],drop_first=True,dtype=int)
```

```python
dataset
```

```python
X=dataset[["Square_Footage","Lot_Size","Garage_Size","Num_Bedrooms_2","Num_Bedrooms_3","Num_Bedrooms_4","Num_Bedrooms_5","Num_Bathrooms_2","Num_Bathrooms_3"]]
y=dataset[["House_Price"]]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train= scaler.fit_transform(X_train)
X_test= scaler.fit_transform(X_test)
```

```python
X_train
```

```python
from sklearn.svm import SVR

model = SVR()  # Create model object
model.fit(X_train, y_train)  # Fit on training data
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
```

```python
r2
```

```python
import pickle
filename="SupportVectorRegression.sav"
pickle.dump(model,open(filename,"wb"))
```
